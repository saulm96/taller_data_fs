```python
import pandas as pd
```

```python
df = pd.read_csv("housing.csv")
df
```

```python
df['calefaccion'] = [1 if isinstance(x, str) and "calefacción" in x.lower() else 0 for x in df['features']]
df['terraza'] = [1 if isinstance(x, str) and "terraza" in x.lower() else 0 for x in df['features']]
df['ascensor'] = [1 if isinstance(x, str) and "ascensor" in x.lower() else 0 for x in df['features']]
df['balcón'] = [1 if isinstance(x, str) and "balcón" in x.lower() else 0 for x in df['features']]
df
```

```python
# Extraer el número de planta
df['planta'] = df['features'].str.extract(r'(\d+)ª Planta')

# Rellenar valores NaN con un valor predeterminado y convertir a entero
df['planta'] = df['planta'].fillna(0).astype(int)

df
```
